```python
import json
from bson import json_util

def transform_dataframe(df, json_col):

    dicts = df[json_col].apply(lambda x: json_util.loads(json_util.dumps(x)))
    json_df = pd.json_normalize(dicts)
    df_reset = df.reset_index(drop=True)
    json_df_reset = json_df.reset_index(drop=True)

    transformed_df = pd.concat([df_reset.drop(columns=[json_col]), json_df_reset], axis=1)

    return transformed_df
```
